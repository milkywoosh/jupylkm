```python
# POD CORESYS

import os
import glob

# path = r'C:\Users\kartu\Downloads\all client'
path = r'C:\Users\kartu\Downloads\all client'
file = glob.glob(os.path.join(path,"*.csv"))

for target in file:
    pod = pd.concat((pd.read_csv(f, dtype={'Tlp1':str, 'No. Referensi': str}, encoding='utf-8', low_memory=False) for f in file))
    
    
    
import copy
df = copy.copy(pod)

df.rename(columns = {' "No. AWB"': 'awb'}, inplace=True)
df['No. Referensi']  = df['No. Referensi'].str.replace('#', '')
df['awb']  = df['awb'].str.replace('#', '')
df['Status POD'].fillna('UNDELIVERED', inplace=True)
# dt=['0000-00-00']
# df = df[~df['Tanggal Terima'].isin(dt)]
df['Status'].fillna('NO Status', inplace=True)
# df['Clientx'] = df['Pelanggan'].str.replace(' ', '')
# df['Clientx'] = df['Clientx'].str.replace('-', '_')
# df['Clientx'] = df['Clientx'].str.slice(0,18)
df['Today'] = pd.to_datetime('now')
# df["Instruksi Khusus"].fillna('atm', inplace=True)


# MERGING with wilayah NOS
df = pd.merge(df, t[['wilayah', 'tlc']], left_on='TLC Tujuan', right_on='tlc', how='left')

df.shape
```

## CEK DELIV UNDEL

```python

def perform_dlvrd(df):
    x = df.groupby(['tlc', 'wilayah','Status POD'])[['awb']].count().unstack()
    z = x.columns
    y = pd.Index([e[1] for e in z.to_list()])
    x.columns = y
    x = x.reset_index()
    x.fillna(0, inplace=True)
    return (x)
perform_dlvrd = perform_dlvrd(df)
```
